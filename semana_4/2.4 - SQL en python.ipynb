**Primero cargar datos...**

```python
# cargar datos

import pandas as pd

df=pd.read_csv('data/shoes.csv')

df.head()
```

```python
# seleccion de datos

df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean', 'upc']]
df.head()
```

```python
# quita nan

df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.sizes=df.sizes.apply(lambda x: x.split(','))
df.colors=df.colors.apply(lambda x: x.split(','))

df.head()
```

```python
# un registro por cada talla y cada color

df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
import warnings
warnings.simplefilter('ignore')

codigo=pd.read_csv('data/upc_corpus.csv')
codigo=codigo.dropna().reset_index(drop=True)

codigo.head()
```

```python
codigo.shape
```

```python
df.ean=codigo.ean.iloc[0:len(df)]
```

```python
# renombrar

df=df.rename(columns={'colors':'color', 
                      'sizes':'size', 
                      'prices.amountMax':'price', 
                      'imageURLs':'url'})

df.head()
```

```python
df['id']=[i for i in range(len(df))]

df=df.drop(columns=['upc'])

df.head()
```

```python
import mysql.connector as conn
```

```python
# crea una base de datos

crea_db=conn.connect(host='localhost',
                     user='root',
                     passwd='password')

cursor=crea_db.cursor()

cursor.execute('drop database if exists productos')
cursor.execute('create database productos')
```

```python
# check

cursor.execute('show databases')

for x in cursor:
    print(x)
```

```python
# conectar otra vez

db=conn.connect(host='localhost',
                     user='root',
                     passwd='password',
                     database='productos')

cursor=db.cursor()
```

```python
# crea una tabla

cursor.execute('drop table if exists zapatos')


tabla='''
            create table zapatos(
                id int,
                brand varchar(50),
                color varchar(50),
                size varchar(50),
                price float,
                url varchar(10000),
                ean varchar(50)
            );
'''


cursor.execute(tabla)
```

```python
# check

cursor.execute('show tables')

for x in cursor:
    print(x)
```

### Inserta datos

```python
%%time

# insert query  (7.02 sec)

for i in range(len(df)):
    
    insert_query='insert into {} ({}) values {};'\
                 .format('zapatos', 
                         ','.join(df.columns), 
                         tuple(df.iloc[i].values))
    
    cursor.execute(insert_query)
    
db.commit()
```
